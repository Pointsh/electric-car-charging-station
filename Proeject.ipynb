```python
import json
import requests
import pandas as pd
import numpy as np
def getLatLng(addr):
    try:
        url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + addr
        headers = {"Authorization": "KakaoAK d83f9b3e940a277cc60f4fbbc850b754"}
        result = json.loads(str(requests.get(url, headers=headers).text))
        match_first = result['documents'][0]['address']
        return float(match_first['y']), float(match_first['x'])
    except:
        match_first='-','-'
        return match_first
```

```python
from pyproj import Proj, transform
proj_UTMK=Proj(init='epsg:5178')
proj_WGS84=Proj(init='epsg:4326')

def transSi(data):
    for i in range(len(data['features'][0]['geometry']['coordinates'][0])):
        text1=data['features'][0]['geometry']['coordinates'][0][i][0]
        text2=data['features'][0]['geometry']['coordinates'][0][i][1]
        tr1,tr2=transform(proj_UTMK,proj_WGS84,text1,text2)
        data['features'][0]['geometry']['coordinates'][0][i][0]=tr1
        data['features'][0]['geometry']['coordinates'][0][i][1]=tr2
    return data
```
